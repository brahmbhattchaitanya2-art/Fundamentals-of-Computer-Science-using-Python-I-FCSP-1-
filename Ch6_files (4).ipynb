```python
try:
    f1=open("mail.txt")
    d={}
    for line in f1:
        if line.startswith("From "):
            print(line)
            w=line.split()[1]
            if w in d:
                d[w]+=1
            else:
                d[w]=1   # here 1  time value store key in d
    print(d)
    m=max(d.values()) # only return the value from dictionary
    for i in d:
        if d[i]==m:   # d [i] is value of that key
            print(i,d[i]) # key and value
            
except Exception as e:
        print("Error:",e)
finally:
    try:
        f1.close()
    except:
        print("No Such fileee")
```
