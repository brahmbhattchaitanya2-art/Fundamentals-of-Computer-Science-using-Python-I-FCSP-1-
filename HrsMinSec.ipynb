```python
def numsq(x,itr):
    guess=x/2
    for i in range(itr):
       guess=(guess+x/guess)/2
    return guess
num=int(input("Enter root"))
guess=int(input("Enter iteration"))
print(numsq(num,guess))
```

```python
def fact(num):
 if(num==0 or num==1):
     return 1
 else:
     return num*fact(num-1)
num=int(input("Enter:"))
print(fact(num))
```

```python
n=int(input("Enter number:"))
if n<=1:
    print("not prime")
else:
    a=2
    while(a<n):
        if(n%2==0):
            print("not prime")
            break
        a=a+1
    else:
            print("prime")
        
```

```python
num=int(input("Enter number:"))
temp=num
l,s=0,0
while temp>0:
    temp=temp//10
    l+=1
temp=num
while temp>0:
    digit=temp%10
    s +=(digit**l)
    temp=temp//10
if s == num:
    print("arm")
else:
    print(" not arm")
    
```
