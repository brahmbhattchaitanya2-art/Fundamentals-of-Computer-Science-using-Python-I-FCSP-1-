```python
e_count = 0
e_sum = 0
e_max = None
e_min = None

o_count = 0
o_sum = 0
o_max = None
o_min = None

while True:
    user = input("Enter Number or 'stop' : ")
    if user == "stop":
        break

    num = int(user)

    if num % 2 == 0:
        e_count += 1
        e_sum += num

        if e_max is None or num > e_max:
            e_max = num
        if e_min is None or num < e_min:
            e_min = num
    else:
        o_count += 1
        o_sum += num

        if o_max is None or num > o_max:
            o_max = num
        if o_min is None or num < o_min:
            o_min = num

if e_count > 0:
    e_avg = e_sum // e_count
    print("For even:", e_max, e_min, e_avg, "(max, min, avg)")
else:
    print("No even numbers")

if o_count > 0:
    o_avg = o_sum // o_count
    print("For odd:", o_max, o_min, o_avg, "(max, min, avg)")
else:
    print("No odd numbers")
    
```

```python


row = int(input("enter row: "))

# upper half
for i in range(row):
    for j in range(i):
        print(" ", end="")        # spaces increase
    for j in range(2 * (row - i) - 1):
        print("*", end="")        # stars decrease
    print()

# lower half
for i in range(1, row):
    for j in range(row - i - 1):
        print(" ", end="")        # spaces decrease
    for j in range(2 * i + 1):
        print("*", end="")        # stars increase
    print()
```

```python
e_count = 0
e_sum = 0
e_max = None
e_min = None

o_count = 0
o_sum = 0
o_max = None
o_min = None

while True:
    user = input("Enter Number or 'stop' : ")
    if user == "stop":
        break

    num = int(user)

    if num % 2 == 0:
        e_count += 1
        e_sum += num

        if e_max is None or num > e_max:
            e_max = num
        if e_min is None or num < e_min:
            e_min = num
    else:
        o_count += 1
        o_sum += num

        if o_max is None or num > o_max:
            o_max = num
        if o_min is None or num < o_min:
            o_min = num

if e_count > 0:
    e_avg = e_sum // e_count
    print("For even:", e_max, e_min, e_avg, "(max, min, avg)")
else:
    print("No even numbers")

if o_count > 0:
    o_avg = o_sum // o_count
    print("For odd:", o_max, o_min, o_avg, "(max, min, avg)")
else:
    print("No odd numbers")
```
