```python
#653
from abc import ABC,abstractmethod

class Shape(ABC):
    @abstractmethod
    def calculate_area(self):
        pass
    
class Rectangle(Shape):
    def __init__(self,l,b):
        self.l=l
        self.b=b
        
    def calculate_area(self):
        self.area_of_rectangle = (self.l * self.b)
        print("Area of rectangle: ",self.area_of_rectangle)
        
    def __gt__(self,o):
        if self.area_of_rectangle >o.area_of_rectangle:
            print("Rect1 has greater area than Rect2")
        else:
            print("Rect2 has greater area than Rect1")
        
class Circle(Shape):
    def __init__(self,r):
        self.r=r
        
    def calculate_area(self):
        self.area_of_circle =(3.14*self.r)**2
        print("Area of circle: ",self.area_of_circle)
    def __gt__(self,ot):
        if self.area_of_circle>ot.area_of_circle:
            print("c1 gt than c2")
        else:
            print("c2 gt than c1")
 
rect1 = Rectangle(10,20)
rect2 = Rectangle(5,2)
rect1.calculate_area()
rect2.calculate_area()
rect1>rect2
print("---------------------------------------------")
c1 = Circle(5)
c1.calculate_area()
c2 = Circle(10)
c2.calculate_area()
c1>c2
print(Circle.__mro__)
```

```python
#651
from abc import ABC, abstractmethod
class Employee(ABC):
    @abstractmethod
    def receive_call(self):
        pass
  
    @abstractmethod
    def end_call(self):
        pass
    @abstractmethod
    def is_free(self):
        pass
  
    @abstractmethod
    def get_rank(self):
        pass
    
class Respondent(Employee):
    def __init__(self,id,name):
        self.id=id
        self.name=name
        self.rank=3
        self.free=True
        
    def receive_call(self):
        print("Call received by",self.name)
        self.free=False
        
    def end_call(self):
        print("Call ended")
        self.free=True
        
    def is_free(self):
        return self.free
    
    def get_rank(self):
        return self.rank
    
class Manager(Employee):
    def __init__(self,id,name):
        self.id=id
        self.name=name
        self.rank=2
        self.free=True
        
    def receive_call(self):
        print("Call received by",self.name)
        self.free=False
        
    def end_call(self):
        print("Call ended")
        self.free=True
        
    def is_free(self):
        return self.free
    
    def get_rank(self):
        return self.rank
    
class Director(Employee):
    def __init__(self,id,name):
        self.id=id
        self.name=name
        self.rank=1
        self.free=True
        
    def receive_call(self):
        print("Call received by",self.name)
        self.free=False
        
    def end_call(self):
        print("Call ended")
        self.free=True
        
    def is_free(self):
        return self.free
    
    def get_rank(self):
        return self.rank
class Call:
    def __init__(self,Id,name):
        self.Id=Id
        self.name=name
        self.assigned = False
    
class CallHandler:
    res=[]
    manager=[]
    director=[]
    
    def add_employee(self,ob):
        if ob.rank==1:
            CallHandler.director.append(ob)
        elif ob.rank==2:
            CallHandler.manager.append(ob)
        elif ob.rank==3:
            CallHandler.res.append(ob)
            
    def dispatch_call(self,call):
        for i in CallHandler.res:
            if i.is_free():
                i.receive_call()
                call.assigned = True
                break
        else:
            for i in CallHandler.manager:
                if i.is_free():
                    i.receive_call()
                    call.assigned = True
                    break
            else:
                for i in CallHandler.director:
                    if i.is_free():
                        i.receive_call()
                        call.assigned = True
                        break
                else:
                    print("Sorry! All employees are currently busy.")
                    
R1 = Respondent(1,"R1")
R2 = Respondent(2,"R2")
R3 = Respondent(3,"R3")

M1 = Manager(11,"M1")
M2 = Manager(12,"M2")

D1 = Director(100,"D1")

ch = CallHandler()
ch.add_employee(R1)
ch.add_employee(R2)
ch.add_employee(R3)
ch.add_employee(M1)
ch.add_employee(M2)
ch.add_employee(D1)

C1 = Call(101,"Ram")
C2 = Call(102,"Rama")
C3 = Call(103,"Aham")
C4 = Call(104,"Alice")
C5 = Call(105,"Zyan")
C6 = Call(106,"Charles")
C7 = Call(107,"Olivia")
C8 = Call(108,"Arjun")


ch.dispatch_call(C1)
ch.dispatch_call(C2)
ch.dispatch_call(C3)
ch.dispatch_call(C4)
ch.dispatch_call(C5)
ch.dispatch_call(C6)
ch.dispatch_call(C7)
ch.dispatch_call(C8)
```

```python
class Vehicle:
    def __init__(self, name, mileage, seatingCapacity):
        self.name = name
        self.mileage = mileage
        self.seatingCapacity = seatingCapacity

    def fare(self):
        return self.seatingCapacity * 100


class Bus(Vehicle):
    def fare(self):
        total_fare = super().fare()          # normal fare
        maintenance = total_fare * 0.10      # 10% extra
        return total_fare + maintenance      # final fare

# -------- Example ----------
bus = Bus("Bus", 10, 50)
car = Vehicle("Car", 15, 5)

print("The bus seating capacity is", bus.seatingCapacity, ". so, the final fare amount should be", bus.fare())
print("The car seating capacity is", car.seatingCapacity, ". so, the final fare amount should be", car.fare())
```
