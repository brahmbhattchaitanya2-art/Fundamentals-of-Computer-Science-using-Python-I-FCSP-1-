```python
# Trivago-style Travel Package Calculator

print("=== Trivago-style Travel Package Calculator ===")

# 1. Input number of travelers
travellers = int(input("Enter number of travelers: "))

# 2. Destination Selection
print("Destinations: 1=Goa (₹20000), 2=Kerala (₹22000), 3=Manali (₹18000), 4=Jaipur (₹15000)")
d = int(input("Enter destination choice (1-4): "))

if d == 1:
    base_price = 20000
    destination = "Goa"
elif d == 2:
    base_price = 22000
    destination = "Kerala"
elif d == 3:
    base_price = 18000
    destination = "Manali"
else:
    base_price = 15000
    destination = "Jaipur"

# 3. Travel Mode
print("Travel Modes: 1=Flight (₹8000), 2=Train AC (₹3000), 3=Train Non-AC (₹1500)")
t = int(input("Enter travel choice (1-3): "))

if t == 1:
    travel_cost = 8000
    travel_mode = "Flight"
elif t == 2:
    travel_cost = 3000
    travel_mode = "Train AC"
else:
    travel_cost = 1500
    travel_mode = "Train Non-AC"

# 4. Hotel Type
print("Hotels: 1=5-Star, 2=3-Star, 3=Budget")
h = int(input("Enter hotel choice (1-3): "))
nights = int(input("Enter number of nights: "))

if h == 1:
    hotel_rate = 5000
    hotel_type = "5-Star"
elif h == 2:
    hotel_rate = 3000
    hotel_type = "3-Star"
else:
    hotel_rate = 1500
    hotel_type = "Budget"

hotel_cost = hotel_rate * nights

# 5. Food Plan
print("Food Plan: 1=Full Meals (₹4000), 2=Breakfast Only (₹2000), 3=None (₹0)")
f = int(input("Enter food plan (1-3): "))

if f == 1:
    food_cost = 4000
    food_plan = "Full Meals"
elif f == 2:
    food_cost = 2000
    food_plan = "Breakfast Only"
else:
    food_cost = 0
    food_plan = "No Meals"

# 6. Sightseeing Add-on
sightseeing = input("Sightseeing Add-on? (Y/N): ")
if sightseeing.upper() == "Y":
    sightseeing_cost = 3500
else:
    sightseeing_cost = 0

# 7. Travel Month (Seasonal Surcharge)
month = input("Travel Month (Jan-Dec): ")
seasonal_surcharge = 0
if month.lower() in ["dec", "may"]:
    seasonal_surcharge = 0.3 * (base_price + travel_cost)

# 8. Per Person Cost before discount
total_before_discount = base_price + travel_cost + hotel_cost + food_cost + sightseeing_cost + seasonal_surcharge

# 9. Discounts
if total_before_discount >= 50000:
    discount = total_before_discount * 0.10
elif total_before_discount >= 40000:
    discount = total_before_discount * 0.07
else:
    discount = 0

# 10. GST (18%)
gst = 0.18 * (total_before_discount - discount)

# 11. Final per-person price
final_per_person = total_before_discount - discount + gst

# 12. Total for all travellers
total_group_cost = final_per_person * travellers

# === OUTPUT ===
print("\n--- Full Package Details ---")
print(f"Destination: {destination}")
print(f"Travel Mode: {travel_mode}")
print(f"Hotel Type: {hotel_type} for {nights} nights")
print(f"Food Plan: {food_plan}")
print(f"Sightseeing: {'Included' if sightseeing.upper() == 'Y' else 'Not Included'}")
print(f"Travel Month: {month}")

print("\n--- Cost Breakdown (Per Person) ---")
print(f"Base Price = {base_price}")
print(f"Travel Cost = {travel_cost}")
print(f"Hotel Cost = {hotel_cost}")
print(f"Food Cost = {food_cost}")
print(f"Sightseeing Cost = {sightseeing_cost}")
print(f"Seasonal Surcharge = {seasonal_surcharge}")
print(f"Total Before Discount = {total_before_discount}")
print(f"Discount = {discount}")
print(f"GST (18%) = {gst}")
print(f"Final Package Price Per Person = {final_per_person}")
print(f"\n=== Total Package Price for {travellers} Travelers = {total_group_cost} ===")
```
