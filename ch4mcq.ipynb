```python
#palindrome
def check(s):
    if len(s) <= 1:
        return True
    else:
        return s[0] == s[-1] and check(s[1: -1])
print(check('saippuakivikauppias'))#vi
print(check('saippuakivkauppias'))#i
```

```python
 def enc(st):
    encoded=""
    c=1
    ld=st[0]
    for i in range (1,len(st)):
        if ld==st[i]:
            c=c+1
        else:
            encoded=encoded+str(c)+ld
            c=0
            ld=st[i]
            c=c+1
    encoded=encoded+str(c)+ld
    return encoded
 st="AAABBACCAA"
 print(enc(st))
```
