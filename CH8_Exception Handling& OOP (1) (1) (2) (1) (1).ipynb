```python
class Demo:
    def _init_(self):
        while True:
            try:
                self.a = int(input("Account: "))
                self.n = input("Enter Name: ")
                self.b = int(input("Balance: "))
                if self.b < 5000:
                    raise Exception("Insufficient (min 5000 required)")
            except Exception as e:
                print(e)
            else:
                break

    def display(self):
        print("Account:", self.a)
        print("Name:", self.n)
        print("Balance:", self.b)

    def credit(self):
        amt = int(input("Enter Credit Amount: "))
        self.b = self.b + amt
        print("Updated Balance:", self.b)

    def debit(self):
        while True:
            try:
                amt = int(input("Enter Debit Amount: "))
                if amt > self.b:
                    raise Exception("Insufficient balance")
                self.b = self.b - amt
            except Exception as e:
                print(e)
            else:
                print("Updated Balance:", self.b)
                break


try:
    ob = Demo()

    user = int(input("1.Display\n2.Credit\n3.Debit\nChoose: "))
    if user == 1:
        ob.display()
    elif user == 2:
        ob.credit()
    elif user == 3:
        ob.debit()
    else:
        print("Invalid choice")

except Exception as e:
    print(e)
    print("Unsuccessful")
else:
    print("Successful")
```

```python
class Student:
    def __init__(self):
        # list will store each student as a dictionary
        self.students = []

    # 1) accept details of n students
    def accept(self):
        n = int(input("How many students? "))
        for _ in range(n):
            name = input("Name: ")
            roll = int(input("RollNo: "))
            m1 = int(input("Marks1: "))
            m2 = int(input("Marks2: "))

            # avoid duplicate roll numbers
            if any(s["roll"] == roll for s in self.students):
                print("Roll number already exists! Student not added.")
            else:
                self.students.append({"name": name, "roll": roll, "m1": m1, "m2": m2})

    # 2) display all students
    def display(self):
        if len(self.students) == 0:
            print("No student records.")
            return

        print("\nList of Students")
        for s in self.students:
            print("Name :", s["name"])
            print("RollNo :", s["roll"])
            print("Marks1 :", s["m1"])
            print("Marks2 :", s["m2"])
            print("-" * 20)

    # 3) search by roll number
    def search(self):
        r = int(input("Enter roll number to search: "))
        for s in self.students:
            if s["roll"] == r:
                print("\nStudent Found")
                print("Name :", s["name"])
                print("RollNo :", s["roll"])
                print("Marks1 :", s["m1"])
                print("Marks2 :", s["m2"])
                return
        print("Student not found.")

    # 4) delete by roll number
    def delete(self):
        r = int(input("Enter roll number to delete: "))
        for i in range(len(self.students)):
            if self.students[i]["roll"] == r:
                del self.students[i]
                print("Deleted successfully.")
                return
        print("Student not found.")

    # 5) update roll number (old -> new)
    def update(self):
        old = int(input("Enter old roll number: "))
        new = int(input("Enter new roll number: "))

        if any(s["roll"] == new for s in self.students):
            print("New roll number already exists! Cannot update.")
            return

        for s in self.students:
            if s["roll"] == old:
                s["roll"] = new
                print("Updated successfully.")
                return

        print("Student not found.")


# -------- main --------
obj = Student()
obj.accept()

while True:
    print("\n1.Display\n2.Search\n3.Delete\n4.Update\n5.Exit")
    ch = int(input("Enter choice: "))

    if ch == 1:
        obj.display()
    elif ch == 2:
        obj.search()
    elif ch == 3:
        obj.delete()
        print("\nList after deletion")
        obj.display()
    elif ch == 4:
        obj.update()
        print("\nList after updation")
        obj.display()
    elif ch == 5:
        break
    else:
        print("Invalid choice.")
        
```

```python
# -------- Custom Exceptions --------
class InvalidToppingError(Exception):
    pass

class InvalidCheeseError(Exception):
    pass

class InvalidSizeError(Exception):
    pass


# -------- Pizza Class --------
class Pizza:
    SIZES = {"small": 50, "medium": 100, "large": 200}

    TOPPINGS = {"corn", "tomato", "onion", "capsicum", "mushroom", "olives", "broccoli"}
    EXOTIC_TOPPINGS = {"broccoli", "olives", "mushroom"}  # 50 each, others 20 each

    CHEESE = {"mozzarella", "feta", "cheddar"}  # 50 each cheese type

    def __init__(self, size, toppings, cheese):
        # size: one
        size = size.strip().lower()
        if size not in Pizza.SIZES:
            raise InvalidSizeError("Invalid size! Choose from: small, medium, large")
        self.size = size

        # toppings: many (list)
        cleaned_toppings = [t.strip().lower() for t in toppings]
        for t in cleaned_toppings:
            if t and t not in Pizza.TOPPINGS:
                raise InvalidToppingError(f"Invalid topping: {t}")
        self.toppings = [t for t in cleaned_toppings if t]  # remove empty

        # cheese: many (list)
        cleaned_cheese = [c.strip().lower() for c in cheese]
        for c in cleaned_cheese:
            if c and c not in Pizza.CHEESE:
                raise InvalidCheeseError(f"Invalid cheese: {c}")
        self.cheese = [c for c in cleaned_cheese if c]

    def price(self):
        total = Pizza.SIZES[self.size]

        # toppings price
        for t in self.toppings:
            if t in Pizza.EXOTIC_TOPPINGS:
                total += 50
            else:
                total += 20

        # cheese price (each type costs 50)
        total += 50 * len(self.cheese)
        return total


# -------- Order Class --------
class Order:
    def __init__(self, name, customerid):
        self.name = name
        self.customerid = customerid
        self.pizzas = []

    def order(self):
        print("\n--- Olly's Pizzas Order ---")
        print("Sizes:", ", ".join(Pizza.SIZES.keys()))
        print("Toppings:", ", ".join(sorted(Pizza.TOPPINGS)))
        print("Cheese:", ", ".join(sorted(Pizza.CHEESE)))

        n = int(input("\nHow many pizzas do you want to order? "))
        for i in range(1, n + 1):
            while True:
                try:
                    print(f"\nPizza {i}")
                    size = input("Enter size (small/medium/large): ")

                    top_in = input("Enter toppings (comma separated) or leave blank: ").strip()
                    toppings = [] if top_in == "" else top_in.split(",")

                    ch_in = input("Enter cheese types (comma separated) or leave blank: ").strip()
                    cheese = [] if ch_in == "" else ch_in.split(",")

                    p = Pizza(size, toppings, cheese)
                    self.pizzas.append(p)
                    print("Pizza added successfully ✅")
                    break
                except (InvalidSizeError, InvalidToppingError, InvalidCheeseError) as e:
                    print("Error:", e)
                    print("Please re-enter this pizza details.\n")
                except ValueError:
                    print("Error: Please enter valid input.\n")

    def bill(self):
        if len(self.pizzas) == 0:
            print("No pizzas ordered.")
            return

        print("\n================= OLLY'S PIZZAS BILL =================")
        print(f"Customer Name : {self.name}")
        print(f"Customer ID   : {self.customerid}")
        print("------------------------------------------------------")

        grand_total = 0
        for idx, p in enumerate(self.pizzas, start=1):
            cost = p.price()
            grand_total += cost

            print(f"\nPizza {idx}")
            print("Size     :", p.size)
            print("Toppings :", ", ".join(p.toppings) if p.toppings else "None")
            print("Cheese   :", ", ".join(p.cheese) if p.cheese else "None")
            print("Price    :", cost)

        print("\n------------------------------------------------------")
        print("Total Pizzas :", len(self.pizzas))
        print("Grand Total  :", grand_total)
        print("======================================================\n")


# -------- Main (Object Creation + Output) --------
try:
    print("Welcome to Olly's Pizzas 🍕")
    name = input("Enter customer name: ")
    cid = input("Enter customer id: ")

    ob = Order(name, cid)
    ob.order()
    ob.bill()
except Exception as e:
    print("Something went wrong:", e)
```

```python
import math

class Point:
    def _init_(self, x, y):
        self.x = x
        self.y = y

    # 1) Distance from origin
    def distance_from_origin(self):
        return round(math.sqrt(self.x*2 + self.y*2), 2)

    # 2) Translate the point
    def translate(self, dx, dy):
        return Point(self.x + dx, self.y + dy)

    # 3) Reflect about x-axis
    def reflect_x(self):
        return Point(self.x, -self.y)

    # 4) Distance between two points
    def distance(self, other):
        return round(math.sqrt((self.x - other.x)*2 + (self.y - other.y)*2), 2)


# -------- Test Case --------
p1 = Point(1, 2)

print("Distance from origin -", p1.distance_from_origin())

p2 = p1.translate(1, 1)
print(f"Translate method - point (1,2) translated by (1,1) increment will be at ({p2.x},{p2.y}) now")

p3 = p2.reflect_x()
print(f"Reflect_x Method - Point (2,3) after given reflection will be at ({p3.x},{p3.y})")

p4 = Point(3, 4)
print(f"Distance Method - distance between point ({p3.x},{p3.y}) and (3,4) is {p3.distance(p4)}")
```
