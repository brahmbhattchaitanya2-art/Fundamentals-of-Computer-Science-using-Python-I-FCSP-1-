```python
u,l,d,s=0,0,0,0
pwd=input("Password:")
if len(pwd)>=8:
    for chr in pwd:
        if chr.isupper():
            u+=1
        elif chr.islower():
            l+=1
        elif chr.isdigit():
            d+=1
        elif chr in "_@$":
            s+=1
if(u and l and d and s):
    print("valid Password")
else:
    print("Invalid Password")
    
```

```python
txt=input("Enter String:")
shift=int(input("shift:"))
ans=" "
for ch in txt:
        if ch.isupper():
            en=ord(ch)+shift
            if en>90:         #Z=90
                en=(en%91)+65   #A=65
            ans+=chr(en)
        elif ch.islower():
            en=ord(ch)+shift
            if en>122:        #z=122
                en=(en%123)+97   #a=97
            ans+=chr(en)
        else:
            ans+=ch
print(ans)
    
    
```
